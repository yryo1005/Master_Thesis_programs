```python
import json
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
with open("generated_ohgiri.json", "r") as f:
    generated_ohgiri = json.load(f)
```

```python
D = np.random.choice(generated_ohgiri)

fig = plt.figure()
ax = fig.add_subplot()
ax.imshow(Image.open(f"../../datas/boke_image/{D['image_number']}.jpg"))
ax.axis("off")

for K, V in D.items():
    print(K, V)
```
